```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Step 1: Download Nifty 50 data (^NSEI is the Yahoo symbol for Nifty 50)
data = yf.download("^NSEI", start="2000-01-01", end="2024-01-01")
data = data[['Close']].dropna()
```

```python
# Step 2: Create feature (Days as X)
data['Days'] = np.arange(len(data)).reshape(-1, 1)
```

```python
# Step 3: Train Linear Regression model
model = LinearRegression()
X = data['Days'].values.reshape(-1, 1)
y = data['Close'].values
model.fit(X, y)
```

```python
# Step 4: Predict prices
data['Predicted_Close'] = model.predict(X)
```

```python
# Step 5: Plot actual vs predicted
plt.figure(figsize=(10, 5))
plt.plot(data['Close'], label="Actual Close Price")
plt.plot(data['Predicted_Close'], label="Predicted (Linear)", linestyle='--')
plt.title("Linear Regression on Nifty 50 Index")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```
