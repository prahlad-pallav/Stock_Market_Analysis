Now I'll fetch the data using the `yfinance` library.

```python
import yfinance as yf

# Fetch Nifty 50 data
nifty50_data = yf.download('^NSEI', period='max')

# Display the first few rows of the data
print(nifty50_data.head())
```

Now, I'll plot the closing price of Nifty 50 over time using `matplotlib`.

```python
import matplotlib.pyplot as plt

# Plot the closing price
plt.figure(figsize=(12, 6))
plt.plot(nifty50_data.index, nifty50_data['Close'])
plt.title('Nifty 50 Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.show()
```

Following the plan, I will now fetch the historical data for HDFC Bank using `yfinance`. The ticker symbol for HDFC Bank on the National Stock Exchange (NSE) is typically 'HDFCBANK.NS'. I will fetch the data for the same period as the Nifty 50 data we already have.

```python
# Fetch HDFC Bank data
# The ticker symbol for HDFC Bank on NSE is 'HDFCBANK.NS'
hdfcbank_data = yf.download('HDFCBANK.NS', period='max')

# Display the first few rows of the HDFC Bank data
print(hdfcbank_data.head())
```

```python
# Plot the 'Close' price
plt.figure(figsize=(12, 6))
plt.plot(hdfcbank_data['Close'], label='HDFC Bank Close Price', color='blue')
plt.title('HDFC Bank Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price (INR)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import pandas as pd

# Download HDFC Bank and Nifty 50 data (adjusted close by default)
hdfc_data = yf.download('HDFCBANK.NS', period='1y')[['Close']]
nifty_data = yf.download('^NSEI', period='1y')[['Close']]

# Convert 'Close' column to Series with proper names
hdfc_close = hdfc_data['Close'].squeeze().rename('HDFC')
nifty_close = nifty_data['Close'].squeeze().rename('NIFTY50')

# Combine and drop missing values
combined_data = pd.concat([hdfc_close, nifty_close], axis=1).dropna()

# Calculate correlation
correlation = combined_data['HDFC'].corr(combined_data['NIFTY50'])
print(f"Correlation between HDFC Bank and Nifty 50: {correlation:.4f}")
```
